```python
#Importing the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Cell 2: Importing the dataset
dataset = pd.read_csv('Reduced_Train.csv')

dataset.head()
```

```python
# Cell 3: Handling missing data with mode

# For categorical columns
dataset['Warehouse_block'].fillna(dataset['Warehouse_block'].mode()[0], inplace=True)
dataset['Mode_of_Shipment'].fillna(dataset['Mode_of_Shipment'].mode()[0], inplace=True)
dataset['Gender'].fillna(dataset['Gender'].mode()[0], inplace=True)
dataset['Reached.on.Time_Y.N'].fillna(dataset['Reached.on.Time_Y.N'].mode()[0], inplace=True)
```

```python
# Cell 4: Distribution of Nominal Variables using Count Plot
# Warehouse_block
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, x='Warehouse_block')
plt.title('Distribution of Warehouse Block')
plt.show()

# Mode_of_Shipment
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, x='Mode_of_Shipment')
plt.title('Distribution of Mode of Shipment')
plt.show()

# Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, x='Gender')
plt.title('Distribution of Gender')
plt.show()

# Reached.on.Time_Y.N
plt.figure(figsize=(8, 6))
sns.countplot(data=dataset, x='Reached.on.Time_Y.N')
plt.title('Distribution of Reached on Time (Yes/No)')
plt.show()
```

Mode of transportation quite evenly distributed with deliveries slightly leaning towards flight
Almost 2/3 of deliveries have been delivered on time. More women than men have ordered from the service.


```python
# Cell 5: Boxplot to Compare Values Within Groups

plt.figure(figsize=(8, 6))
sns.boxplot(data=dataset, x='Gender', y='Weight_in_gms')  
plt.title('Boxplot: Delivery Time by Gender')
plt.show()
```

There doesn't seem to be a significant difference in delivery weight (or time) by gender, although males might have a slightly higher central tendency (median) compared to females.

```python
# Cell 6: Additional Plot (Violin Plot or Swarm Plot)


plt.figure(figsize=(8, 6))
sns.violinplot(data=dataset, x='Mode_of_Shipment', y='Cost_of_the_Product')  
plt.title('Violin Plot: Delivery Time by Mode of Shipment')
plt.show()
```

Ship:

The cost distribution is relatively balanced, with a peak around the median.
Costs are concentrated near the middle range, but there are some higher-cost outliers.
Road:

The cost distribution has a broader spread compared to Ship, indicating variability in costs.
Flight:

The distribution has a shape similar to Road but with slightly less width.
High and low extremes are present but less  than in other modes.
